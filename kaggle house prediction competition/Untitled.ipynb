Next we will drop the columns which contains null values more than 50% 
For then than 50% we will fill them  in the following values
1) For interger or float values we will take the mean of the column and fill it
2) For object or category values we will take the mode of the column and fill it
3)If it is string then we first covert it to object or category and then we will use 2)
